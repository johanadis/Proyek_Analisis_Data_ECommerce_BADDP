## Visualization & Explanatory Analysis




### Pertanyaan 1 : Bagaimana Tren Gross Revenue Bulanan E-commerce Hingga Agustus 2018?  
Bagaimana Tren Pendapatan Kotor E-commerce Sampai Bulan Agustus 2018?

```python
# Filter data hingga 31 Agustus 2018
orders_filtered = main_data[main_data["order_purchase_timestamp"] <= "2018-08-31"].copy()

# Ekstrak tahun-bulan untuk agregasi
orders_filtered["year_month"] = orders_filtered["order_purchase_timestamp"].dt.to_period("M")

# Agregasi total pembayaran per bulan
monthly_payments = orders_filtered.groupby("year_month")["payment_value"].sum()

# Konversi indeks ke string untuk plot
months = monthly_payments.index.astype(str)
values = monthly_payments.values

# Visualisasi tren payment_value per bulan
plt.figure(figsize=(12, 6))
plt.plot(months, values, marker='o', linestyle='-', color='b')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'R${x:,.0f}'))  # Format sumbu Y dalam mata uang BRL
plt.xticks(rotation=45)
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Total Pendapatan (BRL)", fontsize=12)
plt.title("Gross Revenue E-Commerce Sampai Bulan Agustus 2018", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

**Insight:**
- Terjadi peningkatan pendapatan yang signifikan sejak akhir 2016 hingga 2018, menunjukkan pertumbuhan bisnis e-commerce.
- Puncak pendapatan terjadi pada akhir 2017, kemungkinan karena musim belanja akhir tahun.
- Setelah Desember 2017, pendapatan mengalami beberapa penurunan tetapi tetap tinggi dibandingkan tahun sebelumnya.
- Kesimpulan E-commerce berkembang pesat dengan puncak pendapatan di musim liburan, menunjukkan peluang besar dalam strategi promosi berbasis musiman.

### Pertanyaan 2 : Apa produk dengan penjualan tertinggi dan terendah?

```python
sum_order_items_df = main_data.groupby("product_category_name_english")["product_id"].count().reset_index()
sum_order_items_df = sum_order_items_df.rename(columns={"product_id": "products"})
sum_order_items_df = sum_order_items_df.sort_values(by="products", ascending=False)
sum_order_items_df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(22, 10))

colors_high = ["#00C853", "#C0C0C0", "#C0C0C0", "#A9A9A9", "#C0C0C0"]
colors_low = ["#D32F2F", "#C0C0C0", "#C0C0C0", "#C0C0C0", "#C0C0C0"]

# Produk dengan penjualan tertinggi
sns.barplot(x="product_category_name_english", y="products",
            data=sum_order_items_df.head(5), hue="product_category_name_english",
            palette=colors_high, ax=ax[0], legend=False)
ax[0].set_ylabel("Number of Sales")
ax[0].set_xlabel(None)
ax[0].set_title("Produk dengan Penjualan Tertinggi", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelrotation=45, labelsize=12)

# Produk dengan penjualan terendah
sns.barplot(x="product_category_name_english", y="products",
            data=sum_order_items_df.sort_values(by="products", ascending=True).head(5),
            hue="product_category_name_english", palette=colors_low, ax=ax[1], legend=False)
ax[1].set_ylabel("Number of Sales")
ax[1].set_xlabel(None)
ax[1].set_title("Produk dengan Penjualan Terendah", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelrotation=45, labelsize=12)

plt.suptitle("Penjualan Produk Tertinggi dan Terendah", fontsize=20)
plt.tight_layout()
plt.show()
```

### Pertanyaan 3 : Seberapa puas pelanggan terhadap layanan yang kami berikan?

```python
review_scores = main_data['review_score'].value_counts().sort_values(ascending=False)

most_common_score = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(12, 7))
sns.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            palette=["#00C853" if score == most_common_score else "#C0C0C0" for score in review_scores.index],
            legend=False)

plt.title("Rating by customers for service", fontsize=15)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(fontsize=12)
plt.show()
```

**Insight:**
- Grafik di atas menunjukkan bahwa pelanggan sangat puas dengan layanan yang diberikan. Hal ini dibuktikan dengan dominasi rating **5**, yang memiliki jumlah ulasan terbanyak dibandingkan dengan rating lainnya.

### Pertanyaan 4 : Di wilayah mana pelanggan paling banyak terkonsentrasi secara geografis?

```python
bystate_df = (main_data.groupby("customer_state")
              ["customer_id"].nunique()
              .reset_index()
              .rename(columns={"customer_id": "customer_count"}))

max_state = (geo_df.groupby(["geolocation_zip_code_prefix", "geolocation_state"])
             .size()
             .reset_index(name='count')
             .drop_duplicates(subset="geolocation_zip_code_prefix")
             .drop(columns=['count']))

geolocation_silver = (geo_df.groupby(["geolocation_zip_code_prefix", "geolocation_city", "geolocation_state"])
                      [["geolocation_lat", "geolocation_lng"]]
                      .median()
                      .reset_index()
                      .merge(max_state, on=["geolocation_zip_code_prefix", "geolocation_state"], how="inner"))

customers_silver = customers_df.merge(geolocation_silver,
                                      left_on='customer_zip_code_prefix',
                                      right_on='geolocation_zip_code_prefix',
                                      how='inner')

customers_silver.to_csv("geolocation.csv", index=False)

def plot_brazil_map(data):
    brazil = mpimg.imread(urllib.request.urlopen(
        'https://i.pinimg.com/originals/3a/0c/e1/3a0ce18b3c842748c255bc0aa445ad41.jpg'), 'jpg')
    ax = data.plot(kind="scatter", x="geolocation_lng", y="geolocation_lat",
                   figsize=(10,10), alpha=0.3, s=0.3, c='green')
    plt.axis('off')
    plt.imshow(brazil, extent=[-73.98, -33.8, -33.75, 5.4])
    plt.show()
```

```python
plt.figure(figsize=(13, 7))

most_common_state = bystate_df.loc[bystate_df['customer_count'].idxmax(), 'customer_state']
bystate_df = bystate_df.sort_values(by='customer_count', ascending=False)

sns.barplot(x='customer_state', y='customer_count', data=bystate_df,
            hue='customer_state',
            palette=["#00C853" if state == most_common_state else "#C0C0C0" for state in bystate_df['customer_state']],
            legend=False)

plt.title("Number of Customers from Each State", fontsize=15)
plt.xlabel("State")
plt.ylabel("Number of Customers")
plt.xticks(fontsize=10)
plt.show()
```

```python
plot_brazil_map(customers_silver.drop_duplicates(subset='customer_unique_id'))
```

**Insight:**
- Mayoritas pelanggan berasal dari negara bagian di wilayah **tenggara** dan **selatan**.
